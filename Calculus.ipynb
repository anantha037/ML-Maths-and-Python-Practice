```python
import numpy as np
import matplotlib.pyplot as plt

def simple_gradient_descent(learning_rate=0.1, iterations=20):
    x = 10  # Start with an initial value
    history = []  # Store values for visualization

    for _ in range(iterations):
        grad = 2 * x  # Compute gradient (df/dx = 2x)
        x = x - learning_rate * grad  # Update x
        history.append(x)  # Store x values

    return history

# Run gradient descent
x_values = simple_gradient_descent()

# Plot the descent
x_range = np.linspace(-10, 10, 100)
y_range = x_range**2  # Function f(x) = x^2

plt.plot(x_range, y_range, label="f(x) = x^2")
plt.scatter(x_values, [x**2 for x in x_values], color='red', label="Gradient Descent Steps")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Simple Gradient Descent")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # 100 data points
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3x + noise

# Add bias term (x0 = 1 for each sample)
X_b = np.c_[np.ones((100, 1)), X]

# Mean Squared Error Cost Function
def compute_cost(X, y, theta):
    m = len(y)
    cost = (1/(2*m)) * np.sum((X.dot(theta) - y) ** 2)
    return cost

# Gradient Descent Algorithm
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    cost_history = []
    theta_history = [theta.copy()]  # Store theta updates for visualization

    for _ in range(iterations):
        gradient = (1/m) * X.T.dot(X.dot(theta) - y)  # Compute gradient
        theta -= alpha * gradient  # Update theta
        cost_history.append(compute_cost(X, y, theta))
        theta_history.append(theta.copy())  # Store theta update

    return theta, cost_history, theta_history

# Initialize parameters
theta = np.random.randn(2, 1)  # Random initial values
alpha = 0.1  # Learning rate
iterations = 50  # Number of iterations

# Run gradient descent
theta_opt, cost_history, theta_history = gradient_descent(X_b, y, theta, alpha, iterations)

# Plot Cost Function Over Iterations
plt.figure(figsize=(10, 5))
plt.plot(range(iterations), cost_history, 'b-')
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function Optimization using Gradient Descent")
plt.show()

# Visualizing Regression Line Updates
plt.figure(figsize=(10, 5))
plt.scatter(X, y, color="blue", label="Data")

for i in range(0, iterations, 10):  # Plot regression lines at intervals
    y_pred = X_b.dot(theta_history[i])
    plt.plot(X, y_pred, linestyle="dashed", label=f"Iteration {i}")

# Final optimized regression line
y_final = X_b.dot(theta_opt)
plt.plot(X, y_final, color="red", linewidth=2, label="Final Model")

plt.xlabel("X")
plt.ylabel("y")
plt.title("Gradient Descent Optimization Process")
plt.legend()
plt.show()

print("Optimized Parameters:", theta_opt.ravel())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data  # Features (4D)
y = iris.target  # Labels (3 classes)

# Step 1: Standardize the Data (Zero mean, unit variance)
X_mean = np.mean(X, axis=0)  # Compute mean
X_std = np.std(X, axis=0)    # Compute standard deviation
X_normalized = (X - X_mean) / X_std  # Standardize

# Step 2: Compute Covariance Matrix
cov_matrix = np.cov(X_normalized, rowvar=False)

# Step 3: Compute Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 4: Sort Eigenvalues and Eigenvectors (Descending Order)
sorted_indices = np.argsort(eigenvalues)[::-1]  # Get sorted indices
eigenvalues = eigenvalues[sorted_indices]  # Sort eigenvalues
eigenvectors = eigenvectors[:, sorted_indices]  # Sort eigenvectors accordingly

# Step 5: Select Top 3 Principal Components
top_eigenvectors = eigenvectors[:, :3]  # First 3 principal components

# Step 6: Transform Data to New Space
X_pca_manual = np.dot(X_normalized, top_eigenvectors)

# Step 7: 3D Plot of PCA Transformed Data
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with colors based on target labels
scatter = ax.scatter(X_pca_manual[:, 0], X_pca_manual[:, 1], X_pca_manual[:, 2], c=y, cmap='viridis', s=80)

# Labels and Title
ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")
ax.set_title("PCA (Without Built-in Functions) on Iris Dataset (3D Projection)")

# Add legend
legend_labels = iris.target_names
legend = ax.legend(handles=scatter.legend_elements()[0], labels=legend_labels, title="Classes")
ax.add_artist(legend)

# Show the plot
plt.show()

# Print Explained Variance Ratio
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)
print("Explained Variance Ratio:", explained_variance_ratio[:3])
```
